```python
import os
import augly.image as imaugs
import augly.utils as utils
from IPython.display import display

input_img_path = 'Joey.jpg'

# AugLy scale augmentation
input_img = imaugs.scale(input_img_path, factor=0.8)
display(input_img)
```

```python
# Augmentations to the scaled Images!
display(
imaugs.meme_format(
        input_img,text='''HOW YOU DOIN'?''',
        caption_height=75,
        meme_bg_color=(0, 0, 0),
        text_color=(255, 255, 255),
    )
)
```
